```python
from transformers import pipeline
from PIL import Image

# Inicializar o pipeline para conversão de imagem para texto
pipe = pipeline("zero-shot-image-classification", model="openai/clip-vit-base-patch16", device=0)
```

```python
# Processar o frame e obter o texto
frame_pil = Image.open('Fruta-mais-consumida-no-Brasil-banana-tem-espaco-para-crescer-Blog-Buscar-Rural-scaled.jpg')

text = pipe(frame_pil, candidate_labels=['banana','kiwi','melancia','maça'], multi_label=True )
print(text)
```

```python
frame_pil = Image.open('istock-_kiwi_widelg.jpg')

text = pipe(frame_pil, candidate_labels=['kiwi','banana', 'melancia','maçã'], multi_label=True )
print(text)
```
